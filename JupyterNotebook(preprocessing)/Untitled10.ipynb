```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Parking_Violations_Issued_-_Fiscal_Year_2014__August_2013___June_2014_.csv")
```

```python
df.head()
```

```python
df = df[['Summons Number','Plate ID','Registration State','Plate Type','Issue Date','Violation Code','Vehicle Body Type','Vehicle Make','Vehicle Expiration Date','Violation Time','Vehicle Color','Vehicle Year','Latitude','Longitude']]
```

```python
df = df.sample(n=10000)
```

```python
import random
def randpart():
    rand = random.uniform(0, 1)
    if (rand>0.75): 
        return 1
    else :
        return 0
```

```python
lst = [randpart() for i in range(0,df.shape[0])]
```

```python
df2 = pd.DataFrame(lst)
```

```python
df2= df2.rename(columns={'0': 'Alert'})
```

```python
df2.columns = ['Alert']
```

```python
df = df.reset_index()
```

```python
df = df.drop(columns='index')
```

```python
df1 = pd.merge(df, df2, left_index=True, right_index=True)
```

```python
droneid = ['1','2','3','4']
```

```python
random.choice(droneid)
```

```python
df1['DroneID']= ""
```

```python
for index, row in df1.iterrows():
    df1['DroneID'][index] = random.choice(droneid)
```

```python
df1.head()
```

```python
for index, row in df1.iterrows():
    df1['Latitude'][index] = random.uniform(40.5,41)
    df1['Longitude'][index] = random.uniform(-74.5,-74)
```
